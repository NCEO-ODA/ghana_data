# Plots for districts. 

Set the value of `region_where` to have the desired district. For example, here we use `Garu Tempane` (Upper East) as an example. We first reserve a figure with 4 rows sharing the `x` axis (time).

```python
fig, axs = plt.subplots(nrows=4, ncols=1, sharex=True, figsize=(12,12))
axs = axs.flatten()
region_where = "HRname='Garu Tempane'"
year = 2020

for iplot, (product, variable) in enumerate(zip(
                            ["ERA5", "TAMSAT", "TAMSAT", "MODIS"],
                            ["t2m_mean", "precip", "smc_avail_top", "Fpar_500m"])):
    _ = get_one_region_landcover(product, variable, year, region_where, ax=axs[iplot])
fig.tight_layout()
fig.savefig("Garu_Tempane_2020.pdf", dpi=150, bbox_inches="tight")
```

## For regions

For regions, you may run into out of memory errors running this on Binder. You need to do two things:
1. set `region_ds = "GHA_admbndp1_1m_GAUL.shp"` in `get_one_region_landcover`
2. Set `region_where` to be `"HRname='Upper West'"`.


```python
fig, axs = plt.subplots(nrows=4, ncols=1, sharex=True, figsize=(12,12))
axs = axs.flatten()
region_where = "HRname='Upper West'"
year = 2020

for iplot, (product, variable) in enumerate(zip(
                            ["ERA5", "TAMSAT", "TAMSAT", "MODIS"],
                            ["t2m_mean", "precip", "smc_avail_top", "Fpar_500m"])):
    _ = get_one_region_landcover(product, variable, year, region_where,  region_ds = "GHA_admbndp1_1m_GAUL.shp", 
                                 ax=axs[iplot])
```
